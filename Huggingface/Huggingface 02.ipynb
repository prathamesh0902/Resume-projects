```python
from transformers import pipeline, set_seed
from torch import tensor
```

```python
nlp = pipeline("fill-mask", model = 'bert-base-cased')
print(type(nlp.model))
```

```python
from transformers import BertTokenizer

# Load the base uncased tokenizer
tokenizer = BertTokenizer.from_pretrained('google-bert/bert-base-cased')
```

```python
preds = nlp(f"If you don't {nlp.tokenizer.mask_token} at the sign, you will get a ticket")

print("If you don't *** at the sign, you will get a ticket")

for p in preds:
    print(f"Token: {p['token_str']}. Score: {100*p['score']:,.2f}%")
```

```python
from transformers import AutoTokenizer, RobertaTokenizer
tokenizer = AutoTokenizer.from_pretrained("FacebookAI/roberta-base")


nlp2 = pipeline("fill-mask", model = 'roberta-base')

preds = nlp2(f"If you don't {nlp2.tokenizer.mask_token} at the sign, you will get a ticket")

print("If you don't *** at the sign, you will get a ticket")

for p in preds:
    print(f"Token: {p['token_str']}. Score: {100*p['score']:,.2f}%")
```
